# Exploratory Data Analysis

```python
#Distribution of Distance from Home
#Sample 1000 random rows
sampled_df = df.sample(n=1000)

sns.histplot(sampled_df['distance_from_home'], bins=10, kde=True)
plt.title('Distribution of Distance from Home (1000 Random Rows)')
plt.xlabel('Distance from Home')
plt.ylabel('Frequency')
plt.show()
```

The distribution plot of the distance from home shows that most transactions occur within shorter distances from the cardholder's home. There are fewer transactions as the distance increases, suggesting that most cardholders tend to use their cards close to home, which might indicate normal behavior patterns.

```python
#Distribution of Distance from Last Transaction
sampled_df = df.sample(n=1000)

sns.histplot(sampled_df['distance_from_last_transaction'], bins = 15,kde=True)
plt.title('Distribution of Distance from Last Transaction (1000 Random Rows)')
plt.xlabel('Distance from Last Transaction')
plt.ylabel('Frequency')
plt.show()
```

The distribution of the distance from the last transaction indicates that most transactions happen relatively close to the previous transaction. This pattern suggests that cardholders often make multiple purchases in proximity to each other.

# Box Plot

```python
#Boxplot of Ratio to Median Purchase Price by Fraud
sampled_df = df.sample(n=1000)

sns.boxplot(x='fraud', y='ratio_to_median_purchase_price', data=sampled_df)
plt.title('Boxplot of Ratio to Median Purchase Price by Fraud (1000 Random Rows)')
plt.xlabel('Fraud')
plt.ylabel('Ratio to Median Purchase Price')
plt.show()
```

The boxplot shows that the ratio to median purchase price tends to be higher for fraudulent transactions compared to non-fraudulent ones. This suggests that unusually high purchase amounts relative to the median could be a red flag for fraud.

# Count Plot

```python
# Count Plot of Repeat Retailer
sampled_df = df.sample(n=1000)

sns.countplot(x='repeat_retailer', data=sampled_df)
plt.title('Count of Repeat Retailer (1000 Random Rows)')
plt.xlabel('Repeat Retailer')
plt.ylabel('Count')
plt.show()
```

The count plot for repeat retailers indicates that a significant number of transactions occur with repeat retailers. This suggests a common pattern where cardholders frequently shop at the same stores.

```python
# Count Plot of Used Chip
sampled_df = df.sample(n=1000)

sns.countplot(x='used_chip', data=sampled_df)
plt.title('Count of Used Chip (1000 Random Rows)')
plt.xlabel('Used Chip')
plt.ylabel('Count')
plt.show()
```

The count plot for the use of chips shows that there is a notable distribution of transactions involving chip usage. Understanding the distribution helps in assessing the security measures in place during transactions.

```python
# Count Plot of Used Pin Number
sampled_df = df.sample(n=1000)

sns.countplot(x='used_pin_number', data=sampled_df)
plt.title('Count of Used Pin Number (1000 Random Rows)')
plt.xlabel('Used Pin Number')
plt.ylabel('Count')
plt.show()
```

The count plot for the use of PIN numbers reveals that a substantial portion of transactions do not involve PIN usage. The absence of PIN usage in many transactions might highlight areas where additional security could be implemented.

```python
# Count Plot of Online Order
sampled_df = df.sample(n=1000)

sns.countplot(x='online_order', data=sampled_df)
plt.title('Count of Online Order (1000 Random Rows)')
plt.xlabel('Online Order')
plt.ylabel('Count')
plt.show()
```

The count plot for online orders indicates that a significant number of transactions are made online. This is critical for understanding the fraud landscape as online transactions often pose a higher risk of fraud due to the lack of physical verification mechanisms.

```python
# Count Plot of Fraud
sampled_df = df.sample(n=1000)

sns.countplot(x='fraud', data=sampled_df)
plt.title('Count of Fraudulent Transactions (1000 Random Rows)')
plt.xlabel('Fraud')
plt.ylabel('Count')
plt.show()
```

The count plot for fraudulent transactions highlights the imbalance in the dataset, with genuine transactions far outnumbering fraudulent ones. This imbalance suggests the need for appropriate techniques to handle class imbalance during model training and evaluation.

# Correlation Heatmap

```python
sampled_df = df.sample(n=1000)

sns.heatmap(sampled_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix (1000 Random Rows)')
plt.show()
```

The correlation heatmap displays the relationships between different numerical features in the dataset. It shows how certain variables, like distance-related measures, might correlate with each other and with the target variable. This understanding is essential for feature selection and engineering in building predictive models.

# Pairplot

```python
# Pairplot of Selected Features
sampled_df = df.sample(n=1000)

sns.pairplot(sampled_df[['distance_from_home', 'distance_from_last_transaction', 'ratio_to_median_purchase_price', 'fraud']], hue='fraud')
plt.title('Pairplot of Selected Features (1000 Random Rows)')
plt.show()
```

The pairplot for selected features with fraud as the hue illustrates the relationships and distributions of these features, highlighting how fraudulent and genuine transactions are distributed. This visualization helps in identifying potential patterns and separations between fraudulent and non-fraudulent transactions, aiding in the development of effective fraud detection models.
